```python
def categorize(age):
    if age < 13:
        return 'Child'
    elif 13 <= age < 18:
        return 'Teen'
    elif 18 <= age < 35:
        return 'Young Adult'
    elif 35 <= age < 65:
        return 'Adult'
    else:
        return 'Senior'

data['Age_Category'] = data['age'].apply(categorize)

data['Age_Category'] = pd.Categorical(data['Age_Category'], 
                                       categories=['Child', 'Teen', 'Young Adult', 'Adult', 'Senior'], 
                                       ordered=True)

data[['Birth_Year', 'age', 'Age_Category']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

columns = ["Spent_on_Wine", "Spent_on_Fruits", "Spent_on_Meat", 
           "Spent_on_Fish", "Spent_on_Sweets", "Spent_on_Gold"]

for col in columns:
    plt.figure(figsize=(8, 5))
    sns.barplot(data=data, x="Age_Category", y=col, color='pink', ci=None)
    plt.title(f"Average Spending on {col.replace('_', ' ')} by Age Category")
    plt.xlabel("Age Category")
    plt.ylabel("Average Spending")
    plt.tight_layout()
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Columns to include in the pairplot
columns = ["Annual_Income", "Spent_on_Wine", "Spent_on_Fruits", 
           "Spent_on_Meat", "Spent_on_Fish", "Spent_on_Sweets", "Spent_on_Gold"]

# Create the pairplot
plt.figure(figsize=(10, 8))
sns.pairplot(data=data, vars=columns, hue="Age_Category", palette="coolwarm", diag_kind="kde")

# Add title and adjust spacing
plt.suptitle("Pair Plot: Income vs Spending on Different Products by Age Group", y=1.02)
plt.show()
```
